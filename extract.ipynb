```python
import os
import json
import pandas as pd
```

```python
path = "C:/Phonepay_Transaction_insights/pulse/data/aggregated/transaction/country/india/state/"
```

```python
aggregated_state = os.listdir(path)
column = {"States":[],"Year":[],"Quarter":[],"Transaction_Name":[],"Transaction_Count":[],"Transaction_Amount":[],"Transaction_Type":[]}
```

```python
for state in aggregated_state:
    cur_state = os.path.join(path,state)
    current_state = os.listdir(cur_state)
    for year in current_state:
        sta_year = os.path.join(cur_state,year)
        state_year = os.listdir(sta_year)
        for file in state_year:
            file_path = os.path.join(sta_year,file)
            if file.endswith(".json"):
                with open (file_path,"r") as f:
                    data = json.load(f)
                    for i in data["data"]["transactionData"]:
                        name = i["name"]
                        count = i["paymentInstruments"][0]["count"]
                        typ = i["paymentInstruments"][0]["type"]
                        amount =i["paymentInstruments"][0]["amount"]
                        column["Transaction_Name"].append(name)
                        column["Transaction_Count"].append(count)
                        column["Transaction_Type"].append(typ)
                        column["Transaction_Amount"].append(amount)
                        column["States"].append(state)
                        column["Year"].append(year)
                        quarter = int(os.path.splitext(file)[0])
                        column["Quarter"].append(quarter)
aggregated_transaction = pd.DataFrame(column)

aggregated_transaction['States'] = aggregated_transaction['States'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar Islands')
aggregated_transaction['States'] = aggregated_transaction['States'].str.replace('-',' ')
aggregated_transaction['States'] = aggregated_transaction['States'].str.title()
aggregated_transaction['States'] = aggregated_transaction['States'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
```
